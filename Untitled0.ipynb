```
#include <iostream>
#include <vector>
#include <string>
#include <iomanip>
using namespace std;

struct Pelicula {
    string titulo;
    string director;
    int anio;
    string genero;
    float duracion;
    double calificacion;
    string idioma;
    char estado;
};

vector<vector<Pelicula>> coleccion(3, vector<Pelicula>(3));

void inicializarColeccion() {
    coleccion[0][0] = {"El Conjuro", "James Wan", 2013, "Terror", 112, 8.0, "Ingles", 'V'};
    coleccion[0][1] = {"Hereditary", "Ari Aster", 2018, "Terror", 127, 8.5, "Ingles", 'V'};
    coleccion[0][2] = {"Rec", "Jaume Balaguero", 2007, "Terror", 78, 7.5, "Espanol", 'N'};
}

void mostrarColeccion() {
    cout << "\n--- Coleccion de Peliculas de Terror ---\n";
    for (int i = 0; i < coleccion.size(); i++) {
        for (int j = 0; j < coleccion[i].size(); j++) {
            if (!coleccion[i][j].titulo.empty()) {
                cout << setw(15) << coleccion[i][j].titulo << " | "
                     << coleccion[i][j].director << " | "
                     << coleccion[i][j].anio << " | "
                     << coleccion[i][j].duracion << " min | "
                     << coleccion[i][j].calificacion << "/10 | "
                     << coleccion[i][j].idioma << " | "
                     << coleccion[i][j].estado << endl;
            }
        }
    }
}

void buscarPorTitulo() {
    string titulo;
    cout << "Ingrese titulo: ";
    getline(cin, titulo);

    for (auto fila : coleccion) {
        for (auto p : fila) {
            if (p.titulo == titulo) {
                cout << "Encontrada: " << p.titulo
                     << " (" << p.anio << ") - Dir: "
                     << p.director << endl;
                return;
            }
        }
    }
    cout << "Pelicula no encontrada.\n";
}

void modificarPelicula() {
    string titulo;
    cout << "Ingrese titulo a modificar: ";
    getline(cin, titulo);

    for (int i = 0; i < coleccion.size(); i++) {
        for (int j = 0; j < coleccion[i].size(); j++) {
            if (coleccion[i][j].titulo == titulo) {
                cout << "Nuevo director: ";
                getline(cin, coleccion[i][j].director);

                cout << "Nuevo anio: ";
                cin >> coleccion[i][j].anio;
                cin.ignore();

                cout << "Nueva duracion: ";
                cin >> coleccion[i][j].duracion;
                cin.ignore();

                cout << "Nueva calificacion: ";
                cin >> coleccion[i][j].calificacion;
                cin.ignore();

                cout << "Nuevo idioma: ";
                getline(cin, coleccion[i][j].idioma);

                cout << "Nuevo estado (V/N): ";
                cin >> coleccion[i][j].estado;
                cin.ignore();

                cout << "Modificada correctamente.\n";
                return;
            }
        }
    }
    cout << "Pelicula no encontrada.\n";
}

void eliminarPelicula() {
    string titulo;
    cout << "Ingrese titulo a eliminar: ";
    getline(cin, titulo);

    for (int i = 0; i < coleccion.size(); i++) {
        for (int j = 0; j < coleccion[i].size(); j++) {
            if (coleccion[i][j].titulo == titulo) {
                coleccion[i][j] = Pelicula();
                cout << "Pelicula eliminada.\n";
                return;
            }
        }
    }
    cout << "Pelicula no encontrada.\n";
}

void promedioCalificacion() {
    double suma = 0;
    int contador = 0;

    for (auto fila : coleccion) {
        for (auto p : fila) {
            if (!p.titulo.empty()) {
                suma += p.calificacion;
                contador++;
            }
        }
    }

    if (contador > 0)
        cout << "Promedio: " << (suma / contador) << "/10\n";
    else
        cout << "No hay peliculas.\n";
}

void menu() {
    int opcion;

    do {
        cout << "\n--- Menu ---\n";
        cout << "1. Mostrar\n";
        cout << "2. Buscar\n";
        cout << "3. Modificar\n";
        cout << "4. Eliminar\n";
        cout << "5. Promedio\n";
        cout << "6. Salir\n";
        cout << "Opcion: ";
        cin >> opcion;
        cin.ignore(); // IMPORTANTE para Colab

        switch (opcion) {
            case 1: mostrarColeccion(); break;
            case 2: buscarPorTitulo(); break;
            case 3: modificarPelicula(); break;
            case 4: eliminarPelicula(); break;
            case 5: promedioCalificacion(); break;
            case 6: cout << "Saliendo...\n"; break;
            default: cout << "Opcion invalida\n";
        }

    } while (opcion != 6);
}

int main() {
    inicializarColeccion();
    menu();
    return 0;
}
```
